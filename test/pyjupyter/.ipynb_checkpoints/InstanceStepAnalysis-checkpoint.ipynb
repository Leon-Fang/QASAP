```python
import pandas as pd
import math
import numpy as np
from pyecharts.charts import Bar
```

```python
pi=pd.read_csv('instance.csv')
noPi=pi[~pi['InstanceName'].str.contains('run5SerialMockService')]
noPiNO=noPi['InstanceId']
noPiNO
```

```python
df0=pd.read_csv('step.csv')
df1=df0.set_index('InstanceId')
df1.head()
```

```python
df2=df1.drop(df1.loc[noPiNO].index).reset_index()
df = df2.dropna(subset=['StepName']).reset_index()
df
```

```python
dd = df[['StepName','Duration']].reset_index(drop=True)
ds = dd.groupby(['StepName']).describe().reset_index()
ds
```

```python
stepcnt=ds['StepName'].count()
step1 = ds['StepName'][stepcnt-1]
step1data=df.set_index('StepName').loc[step1]
step1data
```

```python
num=math.floor(step1data['Duration'].count()*0.9)-1
step1data['Duration'].sort_values()[num]
```
